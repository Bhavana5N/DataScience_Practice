```python
# . Given the matrix X and the vector y, solve for the vector b in the equation y = Xb using matrix
# inversion and LU factorization.
```

```python
import numpy as np
import scipy.linalg as linalg 
x=np.array([[1,3], [2,1]])
y = np.array([1,7])
print(x)
y
b=np.linalg.solve(x, y )
print(x.dot(b))
print ("Solutions:\n",b )
x_I = np.linalg.inv(x)
print(x_I.dot(y))
```

```python
#LU Factorization
LU = linalg.lu_factor(x) 
y_fac = linalg.lu_solve(LU, y)
print ("Solutions:\n",y_fac)
P, L, U = linalg.lu(x)
print(P)
print(L)
print(U)
```
